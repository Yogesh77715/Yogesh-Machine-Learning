```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=float_cols)
plt.xticks(rotation=90)
plt.title("Box Plot of Float Columns")
plt.show()
```

```python
outlier_counts = {}
for col in float_cols.columns: 
    Q1 = float_cols[col].quantile(0.25)
    Q3 = float_cols[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = ((float_cols[col] < lower_bound) | (float_cols[col] > upper_bound)).sum()
    outlier_counts[col] = outliers
```

```python
outlier_counts_df = pd.DataFrame.from_dict(outlier_counts, orient='index', columns=['Outlier Count'])
print(outlier_counts_df)
```

```python
import numpy as np
def cap_outliers(df):
    for col in df.columns:
        Q1 = df[col].quantile(0.25)
        print(Q1)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
        df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])

    return df
```

```python
float_cols = cap_outliers(float_cols)
```

```python
outlier_counts = {}
for col in float_cols.columns:
    Q1 = float_cols[col].quantile(0.25)
    Q3 = float_cols[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = ((float_cols[col] < lower_bound) | (float_cols[col] > upper_bound)).sum()
    outlier_counts[col] = outliers
```

```python
outlier_counts_df = pd.DataFrame.from_dict(outlier_counts, orient='index', columns=['Outlier Count'])
print(outlier_counts_df)
```

```python
object_cols['diagnosis']=pd.get_dummies(object_cols['diagnosis'],dtype=int, drop_first=True)
```

```python
object_cols
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(float_cols)
X_scaled = pd.DataFrame(X_scaled, columns=float_cols.columns)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data = pd.DataFrame()
vif_data["Feature"] = float_cols.columns
vif_data["VIF"] = [variance_inflation_factor(X_scaled.values, i) for i in range(X_scaled.shape[1])]
vif_data
```

```python
dataset=pd.concat([float_cols,object_cols],axis=1)
```

```python
correlation_with_target = dataset.corr()['diagnosis'].drop('diagnosis')
print(correlation_with_target)
```

```python
from sklearn.model_selection import train_test_split
Y=dataset['diagnosis']
X=dataset.drop(columns='diagnosis')
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn import svm
model=svm.SVR(kernel='rbf')
model.fit(X_train,Y_train)
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5).astype(int)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix

cf=confusion_matrix(Y_test,y_pred)
accuracy = accuracy_score(Y_test, y_pred)
precision = precision_score(Y_test, y_pred)
recall = recall_score(Y_test, y_pred)
print("Confusion Matrix",cf)
print("Accuracy:", accuracy)
print("Precision:",precision)
print("Recall:",recall)
```

```python
from sklearn.metrics import classification_report
report = classification_report(Y_test, y_pred)
print(report)
```
